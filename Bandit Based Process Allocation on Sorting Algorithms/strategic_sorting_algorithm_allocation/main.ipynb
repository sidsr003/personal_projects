```python
from sorting_algorithms import selection_sort, bubble_sort, insertion_sort, merge_sort, quick_sort
from algo_agent import Algo_Agent
import numpy as np
import matplotlib.pyplot as plt
```

```python
algorithms = [selection_sort, bubble_sort, insertion_sort, merge_sort, quick_sort]
num_actions = len(algorithms)
num_tasks = 100
num_runs = 100
initial_action_values = np.zeros(num_actions)
agent = Algo_Agent()

small_arr_means, small_arr_stds, small_arr_visitation_dist = agent.test_allocation(num_actions, initial_action_values, num_tasks, num_runs, 100, algorithms, 0.05)
medium_arr_means, medium_arr_stds, medium_arr_visitation_dist = agent.test_allocation(num_actions, initial_action_values, num_tasks, num_runs, 300, algorithms, 0.05)
large_arr_means, large_arr_stds, large_arr_visitation_dist = agent.test_allocation(num_actions, initial_action_values, num_tasks, num_runs, 700, algorithms, 0.05)
```

```python
axis = plt.subplot(1, 1, 1)

low = small_arr_means-small_arr_stds
high = small_arr_means+small_arr_stds
axis.plot(small_arr_means, label='(array-len=100)', c='deeppink')
axis.fill_between(range(num_tasks), low, high, alpha=0.5, color='coral')

low = medium_arr_means-medium_arr_stds
high = medium_arr_means+medium_arr_stds
axis.plot(medium_arr_means, label='(array-len=300)', c='deeppink')
axis.fill_between(range(num_tasks), low, high, alpha=0.5, color='coral')

low = large_arr_means-large_arr_stds
high = large_arr_means+large_arr_stds
axis.plot(large_arr_means, label='(array-len=600)', c='deeppink')
axis.fill_between(range(num_tasks), low, high, alpha=0.5, color='coral')


axis.set_xlabel('Number of Tasks')
axis.set_ylabel('Execution Time (ms)')
axis.set_title('Sorting Algorithm Allocation')
axis.legend()
```

```python
axis = plt.subplot(1, 1, 1)

bar_width = 0.3
axis.bar(np.arange(num_actions), 100*small_arr_visitation_dist/num_tasks, color='khaki', label='small arr', tick_label=['selection_sort', 'bubble_sort', 'insertion_sort', 'merge_sort', 'quick_sort'], width=bar_width)
axis.bar(np.arange(num_actions)+ 0.5*bar_width, 100*medium_arr_visitation_dist/num_tasks, color='fuchsia', label='medium arr', tick_label=['selection_sort', 'bubble_sort', 'insertion_sort', 'merge_sort', 'quick_sort'], width=bar_width)
axis.bar(np.arange(num_actions)+ 1*bar_width, 100*large_arr_visitation_dist/num_tasks, color='teal', label='large arr', tick_label=['selection_sort', 'bubble_sort', 'insertion_sort', 'merge_sort', 'quick_sort'], width=bar_width)
axis.set_xlabel('Algorithm')
axis.set_ylabel('% Visited')
axis.set_title('Sorting Algorithm Allocation')
axis.legend()
plt.gcf().tight_layout()
```

```python
# For small, medium and large arrays - quicksort is fastest! 
```
