```python
import indicators
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
data = yf.Ticker('TSLA').history(start="2003-01-01", end="2024-01-01", interval="1d")[['Open', 'Close']].to_numpy()
```

```python
axis = plt.subplot()
ema = indicators.ema(data[:, 1], period=26)
macd = indicators.macd(data[:, 1])
rsi = indicators.rsi(data[:, 0], data[:, 1], period=14)

axis.plot(data[:, 0])
axis.plot(data[:, 1])
axis.plot(ema)
axis.plot(macd)
axis.plot(rsi)
```
